Constructing a linear regression model one will, rather sooner than later, come across the F-value.

What is it? What does it know? Does it know things? Let's find out!



F-value is the outcome of the F-test, which is used to check the overall significance of a model.

What it does is really simple and intuitive.

The point of the F-test is to measure the goodness of our model's fit. As usual with linear regression, it is done by using the sum of squares of residuals.
It is then compared with the fit of the model with no predictors (no features). If the sum of squares of residuals in our model is statistically significantly lower, than it means that it is better than the model with no predictors (also often called intercept-only model).

As there are no features in the second model, the only predicting power it has is the mean of the dependent variable, so what we really do is:

we compare OUR fancier, complex model, which uses different features
with the simplest predictive model we can imagine - assigning the mean value to each sample.

The formula for F-value is:
\begin{align*}
F_{stat} = \frac{SS_{mean} - SS_{model}}{SS_{model}}
\end{align*}

```python
import scipy.stats as st
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def find_F_value(x, y):
    slope, intercept, rvalue, pvalue, stderr = st.linregress(x, y)
    SS_mean = 0 #to bdzie suma kwadratw odchyle od sredniej, jakbysmy estymowali tylko srednia wartoscia
    for obs in y:
        SS_mean += (obs-y.mean())**2 
    SS_line = 0 #suma kwadratw odchyle od linii regresji
    for i in range(x.size):
        SS_line += (y[i] - intercept - slope*x[i])**2
    F_stat = (SS_mean - SS_line)/SS_line
    return F_stat

n = 10000 #ilos powtrze operacji
f_set = np.empty(n) #to bdzie zbir rnych wartosci F

for i in range(n):
    x = np.random.choice(range(15), size=10)
    y = np.random.choice(range(15), size=10)
    f_set[i] = find_F_value(x, y)
    
sns.distplot(f_set)

#jesli nasze f z modelu jest gdzies tam w srodku, to mozna uznac ze jest przejawem przypadku
# natomiast jesli nasze f przechodzi dalej w prawo, za wartosc dystrubuanty 95%, wtedy odrzucamy h0 o losowosci na rzecz h1
```
